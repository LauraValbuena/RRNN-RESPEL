# Importar Librerias

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OrdinalEncoder
```

# Filtrado y Limpieza de Bases Anonimizadas RESPEL (2015-2020)

Se cargan los archivos Respel (2015-2020);                                                                                     Se filtra unicamente la corriente de Residuos Peligrosos con codigo A1180 y se consolidan en un DataFrame

```python
Names = ['RESPEL_2015','RESPEL_2016','RESPEL_2017','RESPEL_2018','RESPEL_2019','RESPEL_2020']
df_conso =  pd.DataFrame()

#Se recorren los archivos RESPEL (2015-2020)
for i in range(len(Names)):
    RESPEL = pd.read_excel(Names[i]+'.xlsx')
    print("Data: "+Names[i])
    print(f"Shape of data: {RESPEL.shape}")
    print(f"Number of missing values: {RESPEL.isnull().sum().sum()}")
    ##Filtrar solo CORRIENTE DE RESIDUO O DESECHO PELIGROSO con codigo A1180 
    Filter = RESPEL[RESPEL['CORRIENTE_DE_RESIDUO_O_DESECHO_PELIGROSO'].str.contains('A1180', regex=True, na=True)]
    print(f"Shape of data filtered: {Filter.shape}")
    print("\n")
    ##Concatena la data en un DataFrame
    df_conso = pd.concat([df_conso,Filter], axis=0)

df_conso.tail()  
```

Se eliminan las columnas que no serán utilizadas para la investigación, de acuerdo al mapeo de campos

```python
##Eliminar columnas que no serán utilizadas para la predicción
df_conso = df_conso.drop(columns=['CODIGO_DEPARTAMENTO','CODIGO_MUNICIPIO','NUMERO_DE_EMPLEADOS','CANTIDAD_ALMACENADA_AL_FINAL_DEL_PERIODO_POR_EL_GENERADOR',
       'CANTIDAD_ALMACENADA_AL_INICIO_DEL_PERIODO_POR_EL_GENERADOR',
       'CANTIDAD_TOTAL_ALMACENADA_POR_EL_GENERADOR',
       'CANTIDAD_ENTREGADA_AL_FINAL_DEL_PERIODO_A_TERCEROS',
       'CANTIDAD_ENTREGADA_AL_INICIO_DEL_PERIODO_A_TERCEROS',
       'CANTIDAD_TOTAL_ENTREGADA_A_TERCEROS', 'RAZON_SOCIAL_DEL_TERCERO',
       'CANTIDAD_APROVECHADA_Y.O_VALORIZADA_POR_EL_GENERADOR',
       'TIPO_DE_APROVECHAMIENTO',
       'CANTIDAD_APROVECHADA_Y.O_VALORIZADA_POR_EL_TERCERO',
       'TIPO_DE_APROVECHAMIENTO.1', 'RAZON_SOCIAL_DEL_TERCERO.1',
       'CANTIDAD_TRATADA_POR_EL_GENERADOR', 'TIPO_DE_TRATAMIENTO',
       'CANTIDAD_TRATADA_POR_TERCEROS', 'TIPO_DE_TRATAMIENTO.1',
       'RAZON_SOCIAL_DEL_TERCERO.2', 'CANTIDAD_DISPUESTA_POR_EL_GENERADOR',
       'TIPO_DE_DISPOCISION', 'CANTIDAD_DISPUESTA_POR_TERCEROS',
       'TIPO_DE_DISPOSICION.1', 'RAZON_SOCIAL_DEL_TERCERO.3'])
```

Se renombran algunas columnas y se exporta la base filtrada y unificada

```python
df_conso = df_conso.rename(columns={'ACTIVIDAD_ECONOMICA_CIIU_Rev.4.0_A.C.': 'ACTIVIDAD_ECONOMICA_CIIU'})
df_conso = df_conso.rename(columns={'GENERACION_TOTAL_kg.': 'GENERACION_TOTAL_KG'})
df_conso = df_conso.rename(columns={'ANO': 'ANIO'})
df_conso = df_conso.dropna(subset=["DEPARTAMENTO"])
df_conso["ACTIVIDAD_ECONOMICA_CIIU"].replace('([^0-9]*)','',inplace = True, regex = True)
print(f"Shape of data RESPEL filtered and unified: {df_conso.shape}")
df_conso.to_csv('RESPEL_Filtered_Unified.csv', index=False, sep=';')
df_conso
```

# CATEGORIZAR Y ANALIZAR DATOS (2015 - 2020)

CATEGORIZAR Y ANALIZAR POR DERPARTAMETO  (2015 - 2020)

```python
# Copia del DataFrame Consolidado
datos = df_conso.copy()
datos_dep =  pd.DataFrame()

# Filtrado de las columnas 'DEPARTAMENTO','GENERACION_TOTAL_KG','ANIO'
datos_dep = pd.DataFrame(datos, columns=['DEPARTAMENTO','GENERACION_TOTAL_KG','ANIO'])
datos_dep = datos_dep.groupby(['DEPARTAMENTO', 'ANIO'])['GENERACION_TOTAL_KG'].agg('sum').reset_index()
datos_dep.sort_values(['DEPARTAMENTO'], ascending=[True], inplace=True)

# Se genera un código a cada Departamento
encoder = OrdinalEncoder(categories=[datos_dep['DEPARTAMENTO'].unique()])

# Ajustamos el codificador con la variable DEPARTAMENTO y la transformamos
encoder.fit(datos_dep[["DEPARTAMENTO"]])
datos_dep["IDDEPARTAMENTO"] = encoder.transform(datos_dep[["DEPARTAMENTO"]])
datos_dep = pd.DataFrame(datos_dep, columns=['DEPARTAMENTO','IDDEPARTAMENTO','GENERACION_TOTAL_KG','ANIO'])

# Se ordena por IDDEPARTAMENTO Y ANIO
datos_dep.sort_values(['IDDEPARTAMENTO','ANIO'], ascending=[True, True], inplace=True)
# Se Exporta la base agrupada de Departamentos
datos_dep.to_csv('DEPARTAMENTO_GrpConsolidado.csv', index=False, sep=';')

datos_dep
```

De acuerdo a la base anterior, se realiza la sumatoria de GENERACION_TOTAL_KG por Departameto y se exporta

```python
# Sumatoria de GENERACION_TOTAL_KG por DEPARTAMENTO
datos_dep_sum = datos_dep.copy()
datos_dep_sum = datos_dep_sum.groupby(['DEPARTAMENTO'])['GENERACION_TOTAL_KG'].agg('sum').reset_index()

# Se exporta la Base de Departamento, Agrupada por GENERACION_TOTAL_KG y ANIO
datos_dep_sum.to_csv('DEPARTAMENTO_TOTAL_GrpConsolidado.csv', index=False, sep=';')

# Se Ordena descendentemente por GENERACION_TOTAL_KG para visualizar los departamentos con mayor generación de desechos
datos_dep_sum.sort_values(['GENERACION_TOTAL_KG'], ascending=[False], inplace=True)
datos_dep_sum
```

```python
datos_act =  pd.DataFrame()

# Filtrado de las columnas 'ACTIVIDAD_ECONOMICA_CIIU','GENERACION_TOTAL_KG','ANIO'
datos_act = pd.DataFrame(datos, columns=['ACTIVIDAD_ECONOMICA_CIIU','GENERACION_TOTAL_KG','ANIO'])
datos_act = datos_act.groupby(['ACTIVIDAD_ECONOMICA_CIIU', 'ANIO'])['GENERACION_TOTAL_KG'].agg('sum').reset_index()

# Se ordena por ACTIVIDAD_ECONOMICA_CIIU y ANIO
datos_act.sort_values(['ACTIVIDAD_ECONOMICA_CIIU','ANIO'], ascending=[True, True], inplace=True)
# Se Exporta la base agrupada de Activiad Economica
datos_act.to_csv('ACTIVIDAD_ECONOMICA_GrpConsolidado.csv', index=False, sep=';')
datos_act
```
