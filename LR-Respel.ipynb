```python
# Imports necesarios
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

# PREDICCIÓN DE DATOS CON REGRESIÓN LINEAL

PRONOSTICOS POR DEPARTAMENTO CON REGRESIÓN LINEAL

```python
# Carga de datos
Name = 'DEPARTAMENTO_GrpConsolidado'
df = pd.read_csv(Name+'.csv', sep=';')
df
```

```python
# create a copy of the original data
datos_dep = df.copy()
columns = datos_dep['DEPARTAMENTO'].unique()
iddepto = datos_dep['IDDEPARTAMENTO'].unique()
# Definir Dataframe necesarios.
val_metodo =  pd.DataFrame()
datos = datos_dep.copy()

ANIOS = ['2021','2022','2023','2024','2025']

for i in range(len(iddepto)):
    for ANIO in ANIOS:
        filtered_data = datos[datos['DEPARTAMENTO'] == columns[i]]
        dataX =filtered_data[["ANIO"]]
        X_train = np.array(dataX)
        y_train = filtered_data['GENERACION_TOTAL_KG'].values
        # Creamos el objeto de Regresión Linear
        regr = linear_model.LinearRegression()

        # Entrenamos nuestro modelo
        regr.fit(X_train, y_train)
        
        # Hacemos las predicciones que en definitiva una línea (en este caso, al ser 2D)
        y_pred = regr.predict(X_train)
        val_metodo = val_metodo.append({'DEPARTAMENTO' : columns[i], 'ANIO' : ANIO, 'Coefficients' : regr.coef_,'Mean squared error' : "%.2f" % mean_squared_error(y_train, y_pred), 'Variance score' : "%.2f" % r2_score(y_train, y_pred)}, ignore_index=True)
        #Se realiza la Predicción
        y_Dosmil = regr.predict([[ANIO]])
        if y_Dosmil < 0:
            maxindex = np.array([np.mean(y_train)],dtype=np.ndarray)
            y_Dosmil = maxindex   
        #Agrega el datos al DataFrame
        datos = datos.append({'DEPARTAMENTO' : columns[i],'IDDEPARTAMENTO' : iddepto[i],'ANIO' : ANIO, 'GENERACION_TOTAL_KG' : round(float(y_Dosmil),3)}, ignore_index=True)
        filtered_data = datos.copy()
```

```python
#Se ordena por DEPARTAMENTO y ANIO
datos.sort_values(['DEPARTAMENTO', 'ANIO'], ascending=[True, True], inplace=True)
#Se exporta la predicción
datos.to_csv(Raiz+'DEPARTAMENTO_DatosPronostico_LR.csv', index=False, sep=';')

val_metodo.to_csv(Raiz+'DEPARTAMENTO_Varianza y Error del Metodo_LR.csv', index=False, sep=';')
```
